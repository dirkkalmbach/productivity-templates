## Special Characters in Usernames and Passwords

```python
# read csv with special characters from local file
# if not: read from internet

try:
    df = pd.read_csv("data/spec_char.csv")
except:
    url = "https://support.brightcove.com/special-characters-usernames-and-passwords"
    df = pd.read_html(url)[0] #scrape from wikipedia
```

```python
def convert_html(string_to_encode):
    
    """ðŸ§¦ converts string to 'HTML-save' string by replacing special characters"""
    
    string=""
    
    for letter in string_to_encode:
        
        if letter in list(df["Character"]):
            string = string + df[df["Character"]==letter]["Percent-Encoded"].values[0]
            
        else:
            string = string+letter
            
    return string
```

```python
convert_html("1ze<A<GI0AE4,!6820*7pg")
```
